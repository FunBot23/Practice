```python
userInput = input("Enter the temperature in celsius:")
userInput=float(userInput)
celsius_to_farenheit = (userInput*1.8)+32
print("The temperature in farenheit is:", celsius_to_farenheit)
```

```python
userInput = input("Enter the temperature in farenheit:")
userInput=float(userInput)
farenheit_to_celsius = (userInput-32)*1.8
print("The temperature in celsius is:", farenheit_to_celsius)
```

```python
userInput = input("Enter the temperature :")
if "K" in userInput:
    userInput = userInput.strip("K")
    userInput = float(userInput)
kelvin_to_celsius = userInput - 273.15
print("The temperature in celsius is:", kelvin_to_celsius)
kelvin_to_farenheit = (userInput - 273.15)* 1.8 + 32
print("The temperature in celsius is:", kelvin_to_farenheit)

    
```
